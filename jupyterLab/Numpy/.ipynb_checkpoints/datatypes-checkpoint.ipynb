```python
import numpy as np
```

```python
arr = np.array([1,2,3,6,7])
```

```python
print(arr, arr.dtype)
```

```python
arrFloat = np.array([1.1,1.2,1.3])
```

```python
print(arrFloat,arrFloat.dtype)
```

```python
arrBool = np.array([True,False])
```

```python
print(arrBool,arrBool.dtype)
```

```python
arrCom_num = np.array([1+4j])
arrCom_num2 = np.array([2+4j])

print(np.add(arrCom_num,arrCom_num2))
```

```python
print(arrCom_num,arrCom_num.dtype)
```
