```python
import json
clean_data = []
unique_data = set()
def load_data(filename):
    with open(filename,"r") as f:
        data = json.load(f)
    return data
```

```python
data = load_data("record.json")
print(data)
```

```python
#  cleaning the data 
def clear_data(data):
    text_to_number = {"one" : 1,"two" : 2, "three": 3, "four" : 4, "five" : 5}
    for user in data:
         # cleaning data
         raw_rating = str(user["rating"]).strip().lower()
         if(raw_rating in text_to_number):
             raw_rating = text_to_number[raw_rating]
         user["rating"] = raw_rating
        
         # handling missing value 
         age = user.get("age")
         if(age == None):
             user["age"] = None

         # remove duplicate
         if(user["name"].strip() in unique_data):
             continue

         unique_data.add(user["name"].strip())
         clean_data.append(user)
    return clean_data
        
```

```python
data = clear_data(data)
```

```python
# avg and percentate of the poor ratings

def get_insights(data):
    rating  = 0
    for user in data:
        rating += float(user["rating"])
    print(f"avg rating is {rating/len(data)}")

    poor_rating = 0
    for user in data:
        if(float(user["rating"]) < 3):
            poor_rating += 1 

    print(f"% of the poor rating is {poor_rating/len(data) * 100}")
         

    
```

```python
get_insights(data)
```

```python
def get_recommendation(data):
    recommendation = [];

    for user in data:
        curr_recomm = {}
        curr_recomm["name"] = user["name"]
        if(float(user["rating"]) >=4):
            curr_recomm["brand"] = "Apple"
        else:
            curr_recomm["brand"] = "Samsung"

        recommendation.append(curr_recomm)
    return recommendation

            
    
```

```python
print(get_recommendation(data))
```
