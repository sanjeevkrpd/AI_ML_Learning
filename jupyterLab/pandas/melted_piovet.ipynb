```python
import pandas as pd
```

```python
df = pd.DataFrame({
    "country" : [ "USA","USA","INDIA","INDIA"],
    "year" : [2020,2021,2020,2021],
    "sales" : [100,120,90,110],
    "profit" : [20,25,18,22]
})
```

```python
df
```

```python
melted_df = df.melt(

    id_vars=["country", "year"],
    value_vars=["sales", "profit"],
    var_name="metrics",
    value_name="value",

);

melted_df
```

```python
orginal = melted_df.pivot(
    index = ["country","year"],
    columns = "metrics",
    values = "value",
)
```
