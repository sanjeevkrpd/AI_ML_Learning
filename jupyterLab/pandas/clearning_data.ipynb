```python
# Feature Engineering

df2 = df.copy()

df2


# apply 

df2["tax"] =  df2["income"].apply(lambda x : "20%" if x <= 50000  else  "10%")

gender_map = { "Male" : "M" , "Female" : "F" , "Unknown" : "U"}


df2["gender"] = df2["gender"].map(gender_map)

df2 = df2.assign(new_income  = df["income"] * 1.1)

df2["country"] = df2["country"].replace("USA" , "UA")

df2
```

```python
df2.columns = ["Id", "Name", "Age","Country","Gender","Salary","Tax","New_Salary"]
```

```python
df2
```

```python
sorted_values = df2.sort_values(["Salary", "Age"] , ascending=False)

sorted_values = sorted_values.sort_index()
# sorted_values.reset_index()

sorted_values["Ranking"] = sorted_values["Salary"].rank(ascending = False , method = "dense")

sorted_values
```
